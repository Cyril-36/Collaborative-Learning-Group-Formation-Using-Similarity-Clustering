# 05 - Group Formation and Evaluation

Runs the pipeline in notebook-friendly mode, then inspects group metrics, random-baseline significance, cluster characterization, and group rosters.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import pandas as pd
import plotly.express as px
from src import pipeline

PRESENTATION = "AAA_2014J"
```

```python
result = pipeline.run(
    presentation=PRESENTATION,
    bootstrap_b=2,
    sample_n=300,
    n_jobs=1,
    random_baselines=20,
)
result["winner"][["config_id", "reducer", "clusterer", "k", "bootstrap_ari_mean"]]
```

```python
group_metrics = result["group_metrics"]
group_metrics
```

```python
plot_df = group_metrics.melt(id_vars="strategy", var_name="metric", value_name="value")
fig = px.bar(plot_df, x="metric", y="value", color="strategy", barmode="group", title="Group Metrics")
fig.update_layout(xaxis_tickangle=-35)
fig.show()
```

# 02 - Feature Engineering

Builds the learner-level feature matrix and checks missingness, feature groups, and collaboration-specific signals.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import pandas as pd
import matplotlib.pyplot as plt
from src import features, ingest, preprocess

PRESENTATION = ("AAA", "2014J")
```

```python
tables = ingest.run(*PRESENTATION)
X = features.run(tables)
X.shape, X.head()
```

```python
missing = X.drop(columns=["id_student"]).isna().sum().sort_values(ascending=False)
missing[missing > 0].to_frame("missing_values")
```

```python
ids, X_scaled, imputer, scaler, columns, clean_features = preprocess.preprocess(X)
print("Learners:", len(ids))
print("Modeling features after preprocessing:", len(columns))
columns
```

```python
collab_cols = [c for c in X.columns if "collab" in c or "forum" in c]
X[["id_student", *collab_cols]].head(10)
```

```python
summary_cols = [
    "total_clicks",
    "collaborative_clicks",
    "forum_clicks",
    "live_collab_clicks",
    "collaborative_active_days",
    "collaboration_click_ratio",
    "weighted_score",
]
X[[c for c in summary_cols if c in X.columns]].describe().T
```

```python
corr_cols = [c for c in summary_cols if c in clean_features.columns]
corr = clean_features[corr_cols].corr()
corr.style.background_gradient(cmap="RdBu_r", vmin=-1, vmax=1)
```
